```python
#Question 1
from sqlalchemy import create_engine
from sqlalchemy import MetaData
from sqlalchemy import Table, Column
from sqlalchemy import Integer, String
from sqlalchemy import sql, select, join, desc

engine = create_engine('sqlite:///../world.sqlite')
conn = engine.connect()
DBInfo=MetaData(engine)
```

```python
#selects data of interest from database and orders it by year
country=Table('country', DBInfo, autoload=True)
query=select([country.c.Code, country.c.Name, country.c.Region, country.c.IndepYear])\
            .order_by(desc(country.c.IndepYear))
table = conn.execute(query)

for row in table:
    print(row)
```

```python
#Question 2:
#below is the data I am inserting for the country Monetenegro. includes the continent, region, indepyear, and headofstate
my_insert=country.insert().values(Code='MNE', Name='Montenegro', Continent='Europe', Region='Baltic Countries', IndepYear='2006', HeadOfState='Milo Dukanovic')
result = conn.execute(my_insert)
```

```python
#Question 3:
from sqlalchemy import create_engine
from sqlalchemy import MetaData
from sqlalchemy import Table, Column
from sqlalchemy import Integer, String
from sqlalchemy import sql, select, join, desc

engine = create_engine('sqlite:///../world.sqlite')
conn = engine.connect()
DBInfo=MetaData(engine)

city=Table('city', DBInfo, autoload=True)
query=select([city.c.Name, city.c.CountryCode, city.c.District, city.c.Population]).where(city.c.CountryCode.like('MNE'))
table2 = conn.execute(query)

#Inserts both the cities Podgorica and Niksic
my_insert2=city.insert().values(ID='382210', Name='Podgorica', CountryCode='MNE', District='Podgorica', Population='156000')
result = conn.execute(my_insert2)
my_insert3=city.insert().values(ID='382410', Name='Niksic', CountryCode='MNE', District='Niksic', Population='57000')
result = conn.execute(my_insert3)

for row in table2:
    print(row)  
```

```python
#Question 4:
from sqlalchemy import create_engine
from sqlalchemy import MetaData
from sqlalchemy import Table, Column
from sqlalchemy import Integer, String
from sqlalchemy import sql, select, join, desc
import pandas as pd
import matplotlib.pyplot as plt

engine = create_engine('sqlite:///../world.sqlite')
conn = engine.connect()
DBInfo=MetaData(engine)

#selects data of interest from database
country=Table('country', DBInfo, autoload=True)
query=select([country.c.IndepYear, country.c.LifeExpectancy])

#makes chart
ql=pd.read_sql(query, conn)
ql.plot(x='IndepYear', y='LifeExpectancy', kind='scatter')
plt.show()
```
